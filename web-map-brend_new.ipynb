```python
import pandas as pd
import folium
from shapely import geometry
import geopandas as gpd
```

2. Чтение базового слоя

```python
data_point = gpd.read_file('/Users/macbook/Desktop/брендинг/brend/points.geojson')
data_point.head()
```

3. Создание веб-карты

```python
# перепроецирование слоя для работы с folium

data_point = data_point.to_crs("EPSG:4326")
```

```python
# настройка карты
m = folium.Map(location=[data_point.centroid.y.mean(), data_point.centroid.x.mean()], zoom_start=12, tiles="cartodb positron", control_scale=True)
```

```python
m
```

```python
# создание слоев в карте

import folium.plugins
from folium.plugins import GroupedLayerControl
mcg = folium.FeatureGroup(name="карта района Нагатинский Затон")
m.add_child(mcg)

g0 = folium.plugins.FeatureGroupSubGroup(mcg, "граница района")
m.add_child(g0)

g1 = folium.plugins.FeatureGroupSubGroup(mcg, "символы")
m.add_child(g1)

g2 = folium.plugins.FeatureGroupSubGroup(mcg, "маршрут, связывающий символы района")
m.add_child(g2)

folium.LayerControl(collapsed=False).add_to(m)
```

```python
# добавление слоя с границей города

data_border = gpd.read_file('/Users/macbook/Desktop/брендинг/brend/border.geojson')
data_border = data_border.to_crs("EPSG:4326")

geojson_data = data_border.__geo_interface__
folium.GeoJson(
    geojson_data,
    style_function=lambda feature: {
        'fillColor': 'transparent',  
        'color': 'gray',   
        'weight': 2,     
    }
).add_to(g0)

folium.Marker(
    location=[data_border.centroid.y.mean(), data_border.centroid.x.mean()],
    icon=folium.DivIcon(
        html=f'<div style="font-weight: bold; font-size: 12px; color: gray;">{data_border.iloc[0]["name"]}</div>',
        class_name='map-label'
    )
).add_to(g0)

m
```

```python
# Словарь настроек для разных значений 'name'
icon_config = {
    "Остров Мечты": {
        "color": "red",
        "icon": "star",  # иконка дерева (Font Awesome)
        "prefix": "fa"
    },
    "Станция БКЛ": {
        "color": "blue",
        "icon": "subway",  # иконка дерева (Font Awesome)
        "prefix": "fa"
    },
    "Парк Коломенское": {
        "color": "green",
        "icon": "leaf",  # иконка дерева (Font Awesome)
        "prefix": "fa"
    },
    "Бобровый остров": {
        "color": "darkgreen",
        "icon": "tree",  # иконка дерева (Font Awesome)
        "prefix": "fa"
    },
    "Церковь Вознесения Господня в Коломенском": {
        "color": "lightblue",
        "icon": "church",  # иконка дерева (Font Awesome)
        "prefix": "fa"
    },
    "Велопешеходный мост": {
        "color": "cadetblue",
        "icon": "walking",  # иконка дерева (Font Awesome)
        "prefix": "fa"
    },
    "Царская набережная": {
        "color": "darkblue",
        "icon": "users",  # иконка дерева (Font Awesome)
        "prefix": "fa"
    },
    # Дефолтные настройки для остальных
    "default": {
        "color": "lightgray",
        "icon": "info-sign",
        "prefix": "glyphicon"  # стандартный набор иконок Folium
    }
}

# Проходим по данным с безопасной итерацией
for idx, row in data_point.iterrows():
    # Получаем координаты
    coordinates = [row.geometry.y, row.geometry.x]  # широта, долгота
    
    # Определяем настройки иконки
    if row['name'] in icon_config:
        config = icon_config[row['name']]
    else:
        config = icon_config['default']
    
    # Создаём маркер
    marker = folium.Marker(
        location=coordinates,
        popup=folium.Popup(
            f"<b>Имя:</b> {row['name']}<br>"
            f"<b>Описание:</b> {row['descrip']}",
            max_width=300
        ),
        icon=folium.Icon(
            color=config["color"],
            icon=config["icon"],
            prefix=config["prefix"]
        )
    )
    
    # Добавляем в группу
    marker.add_to(g1)

# Отображаем карту
m
```

```python
# второй тематический слой -  маршрут
from folium.plugins import AntPath, BeautifyIcon

data_file = '/Users/macbook/Desktop/брендинг/brend/route_new.geojson'

route_data = gpd.read_file(data_file)  

locations = route_data[['ycoord', 'xcoord']].values.tolist()

ant_path = AntPath(locations=locations,  reverse=False, color="cadetblue", weight=5, pulse_color="white")
ant_path.add_to(g2)

m
```

4. Дополнительный функционал и сохранение
